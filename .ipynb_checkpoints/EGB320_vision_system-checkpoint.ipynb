```python
from collections import namedtuple
from abc import ABC, abstractmethod

class DetectionModel(ABC):
    calibration = None
    

class ThreshBlob(DetectonModel):
    
    EXAMPLE_CALIBRATION = {
        '2d_circularity': None,
        'colorspace_thresholds': None
    }
    
    def __init__(self, calibration=EXAMPLE_CALIBRATION):
        self.calibration = calibration
        super().__init__()
        

class VisualObject():
        
    # real_width <float>
    # real width of object needing detection in metres
    real_width = None
        
    # real_length <float>
    # real length of object needing detection in metres
    real_length = None
        
    # real_height <float>
    # real height of object needing detection in metres
    real_height = None
    
    # detection_model <DetectionModel>
    # the model to be used by the vision system for detection of this object
    detection_model = None
    
    # bearing <float>
    # in radians estimated from the camera's center of translation (for convenience),
    # with the center of vision located at pi radians.
    bearing = None
    
    # distance <float>
    # estimated distance of obect in the previous frame
    distance = None
    
    def __init__(self, real_width, real_length, real_height,  detection_model):
        self.detection_model = detection_model
        
    def update_with_frame(self, frame):
        pass
    

class Odometry():
    dx = None
    dy = None
    dz = None
    
    def update_with_frame(self, frame):
        pass
        

class VisionSystem():
    
    objects_to_track = {
        # 'Ball': VisualObject()
    }
    
    odometry = Odometry()
    
    def __init__(self, objects_to_track):
        self.objects_to_track = objects_to_track
    
    def update_with_frame(self, frame):
        for obj in objects_to_track:
            obj.update_with_frame(frame)
        
        odometry.update_with_frame(frame)


```
